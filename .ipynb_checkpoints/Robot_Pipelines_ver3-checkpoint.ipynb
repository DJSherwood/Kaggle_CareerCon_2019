## Import Libraries

```python
import gc
import os
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
import logging
import datetime
import warnings
import numpy as np
import pandas as pd
import math
from subprocess import check_output
from sklearn.compose import ColumnTransformer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectPercentile, mutual_info_regression
from sklearn.impute import SimpleImputer 
from sklearn.impute import MissingIndicator
from sklearn.linear_model import RidgeClassifier
from sklearn.metrics import mean_squared_error, confusion_matrix
from sklearn.metrics import roc_auc_score, roc_curve, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, cross_val_score
from sklearn.model_selection import StratifiedKFold, GroupKFold, GroupShuffleSplit
#from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import FeatureUnion
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import reciprocal, uniform, kurtosis, skew
import shap
#warnings.filterwarnings('ignore')

train = pd.read_csv('X_train.csv')
test = pd.read_csv('X_test.csv')
y = pd.read_csv('y_train.csv')
```

## Join Data

```python
# Full Dataset
train = train.set_index('series_id').join(y.set_index('series_id'))

# Smaller Dataset
trainb = train.loc[train['group_id'].isin([2,13,23,37,49])]
```

## Create Response

```python
# Label encode the response
encoded_response=pd.DataFrame()
instance_LabelEncoder = LabelEncoder()
encoded_response['surface'] = instance_LabelEncoder.fit_transform(trainb['surface'])
y_data = encoded_response['surface']
```

## Pipeline

```python
# Parameters
parameters = {
    'max_depth': [12,15],
    'n_estimators': [5, 10, 15],
    'min_samples_leaf': [2,4],
    'random_state': [42]
}

# Pipeline
class_pipe = Pipeline([
    ('smote', SMOTE(random_state=42, sampling_strategy='minority')),
    ('add_features', PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)),
    #('standardize', StandardScaler()),
    ('GrdSrch', GridSearchCV(RandomForestClassifier(), param_grid=parameters, scoring='accuracy'))
])
```

```python
# Select few variables; set grouping variable
X_data = trainb[['orientation_X','orientation_Y','orientation_Z','orientation_W','angular_velocity_X','angular_velocity_Y','angular_velocity_Z','linear_acceleration_X','linear_acceleration_Y','linear_acceleration_Z']]
grp_data = trainb['group_id']
```

```python
name = ["Random Forest Classifier"]
model_list = {}

#GrdSrch = GridSearchCV(class_pipe, parameters, verbose=2, scoring='accuracy')
gs_classif = class_pipe.fit(X_data, y_data)
    
# Use cross_val_score
score = cross_val_score(gs_classif, X=X_data, y=y_data, cv=GroupShuffleSplit().split(X=X_data, y=y_data, groups=grp_data))
print("{} score: {}".format(name, score.mean()))
```

```python
gs_classif.named_steps['GrdSrch'].best_estimator_
```

```python
# Run Randomforest outside of pipeline
small_pipe = Pipeline([
    ('add_features', PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)),
  #  ('standardize', StandardScaler())
])

# transform signal data
X_prepared = small_pipe.fit_transform(X_data)

# create classifier
rf = RandomForestClassifier(
    max_depth=gs_classif.named_steps['GrdSrch'].best_params_['max_depth'], 
    min_samples_leaf=gs_classif.named_steps['GrdSrch'].best_params_['min_samples_leaf'], 
    n_estimators=gs_classif.named_steps['GrdSrch'].best_params_['n_estimators'], 
    min_samples_split=2,
    random_state = gs_classif.named_steps['GrdSrch'].best_params_['random_state']
)

# Fit 'best' model on all data
model = rf.fit(X=X_prepared, y=y_data)
```

## Create DataFrame for SHAP

```python
# Convert data to pandas dataframe
data_names = small_pipe.named_steps['add_features'].get_feature_names()
X_prep_df = pd.DataFrame(data=X_prepared, columns=data_names, index=[i for i in range(0, 28800)])
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_prep_df)
```

## SHAP

The SHAP package does two things. First, it generates Shapely values from our predictions. In this case, they are only trivially useful because I used the same dataset that I trained the random forest on. Second, the package creates a model on the Shapely values...I understand this to be an interpolation of sorts. I am given to understand that this is performed to speed up computation time. 

The below graph is the 'default' summary graph. I used a 'tree explainer' because I had used a decision tree to model the data in the first place. Shapley values are calculated by examinging changes in the prediction relative to a cohort of features. The below graph shows that x2 (or 'orientation_z') has the highest average impact for each of the three classes. 

The combination 'x1 x3' ( 'orientation_y * orientation_W' ) is also very high. 

It seems that not all classes have the same 'model output magnitude.' There is very little of 'Class 1' represented by the shapley values of 'orientation_y' by itself. 

In the forums, several kagglers put forth some good a ideas as to why the orientation variables are suspect. Indeed, the below graph shows that *every* feature implicates one of the orientation sensors. What happens if we remove them? 

```python
shap.summary_plot(shap_values, X_prep_df)
```

## Re-run pipe without orientation sensors / features

```python
# Note to self...make this entire thing a class or something to facilitate re-running
# Parameters
parameters = {
    'max_depth': [12,15],
    'n_estimators': [5, 10, 15],
    'min_samples_leaf': [2,4],
    'random_state': [42]
}

# Pipeline - using degree 3 for PolynomialFeatures() this time
class_pipe = Pipeline([
    ('smote', SMOTE(random_state=42, sampling_strategy='minority')),
    ('add_features', PolynomialFeatures(degree=3, interaction_only=True, include_bias=False)),
    ('GrdSrch', GridSearchCV(RandomForestClassifier(), param_grid=parameters, scoring='accuracy'))
])

# Select few variables; set grouping variable
X_data = trainb[['angular_velocity_X','angular_velocity_Y','angular_velocity_Z','linear_acceleration_X','linear_acceleration_Y','linear_acceleration_Z']]
grp_data = trainb['group_id']

name = ["Random Forest Classifier"]
model_list = {}

#GrdSrch = GridSearchCV(class_pipe, parameters, verbose=2, scoring='accuracy')
gs_classif = class_pipe.fit(X_data, y_data)
    
# Use cross_val_score
score = cross_val_score(gs_classif, X=X_data, y=y_data, cv=GroupShuffleSplit().split(X=X_data, y=y_data, groups=grp_data))
print("{} score: {}".format(name, score.mean()))
```

```python
# Run Randomforest outside of pipeline
small_pipe = Pipeline([
    ('add_features', PolynomialFeatures(degree=3, interaction_only=True, include_bias=False)),
])

# transform signal data
X_prepared = small_pipe.fit_transform(X_data)

# create classifier
rf = RandomForestClassifier(
    max_depth=gs_classif.named_steps['GrdSrch'].best_params_['max_depth'], 
    min_samples_leaf=gs_classif.named_steps['GrdSrch'].best_params_['min_samples_leaf'], 
    n_estimators=gs_classif.named_steps['GrdSrch'].best_params_['n_estimators'], 
    min_samples_split=2,
    random_state = gs_classif.named_steps['GrdSrch'].best_params_['random_state']
)

# Fit 'best' model on all data
model = rf.fit(X=X_prepared, y=y_data)
```

#### and then redo the shap explainer

```python
# Convert data to pandas dataframe
data_names = small_pipe.named_steps['add_features'].get_feature_names()
X_prep_df = pd.DataFrame(data=X_prepared, columns=data_names, index=[i for i in range(0, 28800)])

# Explainer
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_prep_df)
```

Now we see a more even representation of our classes...at the cost of being less predictive overall. Remember, 'x2 x5' is 'linear velocity z * linear acceleration z'. Without using orientation this seems almost impossible! 

```python
shap.summary_plot(shap_values, X_prep_df)
```

This looks like a beautiful piece of work. The 'shap_values[0]' means that we are focusing on the predictions for the first class. The "text" for the first argument specifies our feature. And it looks like the dependen_plot automatically selects the feature that is "next down." 

But what does it mean? The color bar specifies the range of values for x2 ('angular velocity Z'). The *x-axis* represents the range of values of 'linear velocity z * linear acceleration z' and the *y-axis* represents how 'linear velocity z * linear acceleration z' changes the value of the prediction. 

The first thing I notice is that negative values less than -2 of 'linear velocity z * linear acceleration z' have negative shapley values, and all values of 'linear velocity z * linear acceleration z' have positive shapley values. Interestingly, *postive values* of 'linear velocity z' (x2) are associated with *negative* values of 'linear velocity z * linear acceleration z'. 

```python
shap.dependence_plot("x2 x5", shap_values[0], X_prep_df)
```

```python
# less confusing
shap.dependence_plot("x2", shap_values[0], X_prep_df)
```
